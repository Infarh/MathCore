```python
import numpy as np
from numpy import pi, sin, cos
import pandas as pd
import matplotlib.pyplot as plt
```

```python
deg = pi / 180
```

```python
f0 = 300
phi0 = 30*deg

def s(t): return sin(2 * pi * f0 * t + phi0)
def sinc(x): return 0 if x == 0 else sin(x) / x
```

```python
N0 = 100
N = 10
fd = 1000
dt = 1 / fd
tt0 = np.linspace(0, (N-1) * dt, N0)
t = np.arange(0, N * dt, dt)
t
```

```python
ss0 = s(tt0)
ss = s(t)
ss
```

```python
plt.plot(tt0, ss0, '--r')
plt.stem(t, ss)
plt.grid();
```
